```python
import pickle

# Create a Python object
my_object = {'name': 'John', 'age': 30}

# Serialize the object and store it in a variable
serialized_data = pickle.dumps(my_object)

# Now, `serialized_data` contains the serialized object
print(serialized_data)  # This will print the serialized data as bytes

# You can also convert it to a string if needed
serialized_data_str = serialized_data.decode('latin1')
print(serialized_data_str)  # This will print the serialized data as a string
```

```python
class Object:
    def __init__(self, name, umur):
        self.name = name
        self.umur = umur

objects = []
while True:
    name = input("Enter the name of the object: ")
    umur = input("Enter the umur of the object: ")
    objects.append(Object(name, umur))
    another = input("Do you want to make another object? (yes/no): ")
    if another.lower() != "yes":
        break

# Now, 'objects' is a list of all the objects created by the user
print(objects)
for obj in objects:
    print(obj.name)
    print(obj.umur)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# No 1

# Membuat larik x yang mencakup rentang yang luas
x = np.linspace(0, 7, 400)
# Menghitung nilai fungsi kuadrat
y = x**2 - 7*x + 12

# Membuat grafik
plt.plot(x, y, label='y = x^2 - 7x + 12')
plt.scatter([4, 3], [0, 0], color='red', marker='o', label='Akar-akar')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

# Memberi label sumbu x dan y
plt.xlabel('x')
plt.ylabel('y')

# Menambahkan label akar-akar
plt.text(4, 0.5, 'x = 4', fontsize=12, ha='center')
plt.text(3, 0.5, 'x = 3', fontsize=12, ha='center')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.grid()
plt.show()
```
